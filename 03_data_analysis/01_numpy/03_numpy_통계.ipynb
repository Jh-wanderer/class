# 통계관련 Numpy

## 중심 경향 측도

```python
import numpy as np
```

```python
scores = np.array([85, 90, 92, 88, 90, 75, 88, 95, 92, 85, 80, 95, 100, 92, 85])

# 평균 mean
print(np.sum(scores) / scores.shape[0])
np.mean(scores)
```

```python
# 최빈값
# 빈도수가 가장 많은 값
values, count = np.unique(scores, return_counts=True)
values, count

max_count = np.max(count)
values[count == max_count]
```

```python
# defaultdict 타입별 기본값을 초기화하는 dict
from collections import defaultdict
count_dict = defaultdict(int) # 요소의 기본값은 0(int기본값)으로 자동설정

for score in scores:
    #key : 점수,  value : 빈도수
    count_dict[score] += 1
print(count_dict)
```

## 산포 측도

```python
# 중위값 median
# 데이터를 크기에 따라 순서대로 정렬했을 때 가운데 값
print(np.sort(scores))
# 짝수면 가운데 두 수의 평균값이 중위값이 됨
np.median(scores)
```

```python
# 분산 varianve
# 평균으로부터 얼마나 떨어져 있는지를 나타내는 지표
# 이 값이 클수록 데이터가 많이 퍼져 있음을 의미한다.
# (값 - 평균)^2 의 평균
print(np.mean((scores - np.mean(scores)) ** 2))
np.var(scores)
```

```python
# 표준편차 standard deviation
# 분산의 제곱근으로, 데이터의 퍼짐정도
np.std(scores)
```

```python
# 범위 계산
# 최대값 - 최소값을 통해 값의 범위를 계산
np.max(scores) - np.min(scores)
```

## 확률 분포
- 정규분포 : 가우시안 표준 정규분포. 평균 기준으로 종모양 배치
    - np.random.randn()
    - np.random.normal()
- 균등분포 : 최소-최대범위안에 값이 균일하게 분포
    - np.random.rand()
    - np.random.randint()


```python
import matplotlib.pyplot as plt
```

```python
# np.random.randn(shape)
# - 평균이 0, 표준편차가 1인 난수 출력 함수
np.random.randn()
```

```python
arr = np.random.randn(10, 10)
arr
```

```python
arr = np.random.randn(1000)

# 히스토그램 (도수분포)
plt.hist(arr, bins=100) # bins 막대개수
plt.show()
```

```python
# np.random.normal(loc, scale, size)
# 정규분포에서 난수를 반환하는 함수
# - loc 평균
# - scale 표준편차
# - size 개수, ndarray.shape

arr = np.random.normal(10, 1, 1000)
print(arr)
plt.hist(arr, bins=100)
plt.show()
```

```python
# np.random.rand()
# 0.0이상 ~ 1.0 미만 사이에서 균등하게 난수를 추출
arr = np.random.rand(1000)
arr

plt.hist(arr, bins=100)
plt.show()
```

```python
np.random.rand(1000)

np.random.rand(2, 3)

arr = np.random.rand(2, 3, 4)
print(arr.shape)
arr
```

```python
# np.random.randint(start, end, size)
# - start이상 end미만에 size개의 균등분포 난수(정수)를 추출
np.random.seed(0) # 난수계산 시작값 seed 고정

arr = np.random.randint(1, 100, 1000)

print(arr)

plt.hist(arr, bins=100)
plt.show()
```
