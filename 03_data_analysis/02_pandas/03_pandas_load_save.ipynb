## 실습문제 - titanic 승객 데이터

**컬럼 설명**
- PassangerId : 승객 고유번호
- Survived : 생존 유무, target 값. 0 = 사망
- Pclass : 객실 등급, 1 / 2 / 3
- Name: 탑승객 성명
- Sex : male / female
- Age : 나이
- Sibsp : 함께 탑승한 형제자매, 배우자 수 총합
- parch : 함께 탑승한 부모, 자녀 수 총합
- ticket : 티켓 넘버
- Fare : 탑승 요금
- Cabin : 객실 넘버
- Embarked : 탑승 항구 (C: cherbourg / Q : queenstown / S : southhampton )

**03. 1등급(Pclass == 1) 승객 중에서 운임(Fare)이 가장 높은 승객의 이름을 조회하라.**
- 1등급 객실에 있는 승객 중에서 운임이 가장 높은 승객의 이름을 조회



```python
df['Pclass'].value_counts()
pclass1_df = df[df['Pclass'] == 1]
pclass1_df_sorted = pclass1_df.sort_values(by ='Fare', ascending=False)
pclass1_df_sorted['Fare Rank'] = pclass1_df_sorted['Fare'].rank(ascending=False, method = 'dense').astype(int)

# idxmax() 해당 컬럼 값 중 최대값을 가진 행의 인덱스를 반환
pclass1_df_sorted.loc[pclass1_df_sorted['Fare'].idxmax()]['Name']

# method = 'average' rank 순위 부여시에 동일한 값이 있을 경우 처리
# - average | min | max | first | dense
```

```python
pclass_1 = df.filter(items=['Pclass' == 1, 'Fare', 'Name'])
pclass_1.sort_values(by=['Fare'], ascending=False)
pclass_1_1 = pclass_1.sort_values(by=['Fare'], ascending=False)

i =
pclass_1_1.head(i)
```
