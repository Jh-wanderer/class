## Figure | Axes
- Figure : 전체 그림이 그려지는 컨테이너
- Axes : 그래프가 실제로 그려지는 영역(좌표, 데이터)
    - Line2D
    - Text
    - XAxis
    - YAxis

```python
# Figure 객체 생성
# - figsize 전체 사이즈 지정
fig = plt.figure(figsize=(6,4), facecolor='silver')

# 현재 Axes 객체를 가져오기
ax = plt.gca() # 축을 가져오는 함수, get current axis
ax.plot([1,2,3],[4,5,6])
print(ax)

plt.title('Hello Mathplotlib')
plt.xlabel('X')
plt.ylabel('Y')

plt.show() # 이게 출력 기준
```

```python
# 여러개의 Axes 객체 작성하기
fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(10,6), facecolor='silver')
print(fig)
print(ax)
```

```python
plt.figure(figsize = (5,3))
plt.show()
# 1. list
# 2. ndarray
x = np.arange(1,11)
y = x * 2
print(x)
print(y)

plt.plot(x,y)
plt.show()

# 3. series
plt.figure(figsize = (3,2))
plt.plot(figsize = (3,2))
ser_x = pd.Series(x)
ser_y = pd.Series(y)
print(ser_x)
print(ser_y)


plt.plot(ser_x, ser_y)
plt.show()

# 4. DataFrame
plt.figure(figsize = (3,2))
(df_x,df_y) = pd.DataFrame({'x':x, 'y':y})
plt.plot(df_x,df_y)
plt.show()
```

```python
# plot 매개변수
fig = plt.figure(figsize = (6,4))
plt.plot(
    x,
    y,
    color = 'blue',
    marker = 'o',
    markersize = 9,
    linestyle = '--',
    linewidth = 2
)
```

```python
# label : 기본적으로 한글 폰트 지원하지 않음. 축의 위험
plt.plot(x,y)
plt.xlabel('시간')
plt.ylabel('Score')
plt.title('Title')
plt.show()
```

```python
# ticks
plt.figure(figsize = (4,8))

x = np.arange(1,101)
y = np.linspace(2, 200, 100)
print(x,y)
```

```python
# 눈금에 대한 설정
plt.xticks(ticks=np.arange(0,101,5), rotation=40) # x축 값들이 겹쳐보이지 않게 rotation
plt.yticks(rotation=40)
plt.plot(x,y)
plt.show()
```

```python
# lim : x축/y축의 보여지는 범위 재지정
# ticks
plt.figure(figsize = (4,8))

x = np.arange(1,101)
y = np.linspace(2,200, 100)

# xlim, ylim : 각 축의 범위를 수동으로 직접 지정
plt.xlim(1,50)
plt.ylim(2,100)
plt.plot(x,y)
plt.show()
```

```python
# legend 범례
plt.figure(figsize = (3,2))
x= np.arange(1,101)
y = np.linspace(2,200,100)
y2 = np.linspace(3,300,100)

plt.plot(x,y, label = '시간')
plt.plot(x,y2, label = '공간')

# legend : plot(label)에 대한 범례
# loc : 범례위치 (best:기본값)
# - (upper|lower) (left|right)
# - center

plt.legend()

plt.show()
```

```python
# 여러 그래프 동시에 그리기
plt.figure(figsize = (6,4))

x = np.arange(1,10)
y1 = x * 2
y2 = x * 3

plt.plot(x,y1, label = '선', color = 'red', marker = 'o', markersize = 10)
plt.plot(x,y2, label = '바', color = 'blue')
plt.show()
```

```python
# subplot 이용해서 여러 그래프를 각각 그리기
x1 = np.arange(1,11)
y1 = x1 * 2
x2 = np.arange(11,20)
y2 = x2 * 2

fig, ax = plt.subplots(1,2, figsize = (10,5)) # 두 그래프 그려지는 틀 지정

# ax[0] plot, xlabel, ylabel, title, legend
ax[0].plot(x1,y1, label='선', color = 'red')
ax[0].set_xlabel('X1') # subplot으로 그릴 때에는 xlabel 지정시 set을 앞에 붙여줘야 함.
ax[0].set_ylabel('Y1')
ax[0].set_title('빨간 선')
ax[0].legend()
ax[0].grid()
plt.show()

ax[1].bar(x2,y2, label='바', color = 'yellowgreen')
ax[1].set_xlabel('X2')
ax[1].set_ylabel('Y2')
ax[1].set_title('연두색 막대')
ax[1].legend()
ax[1].grid()
plt.show()
```
