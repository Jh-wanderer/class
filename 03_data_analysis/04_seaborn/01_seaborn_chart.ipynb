## bar plot
- x축 : 범주형 데이터
- y축 : 연속형 데이터에 대한 집계처리 (평균/총합)


```python
sns.barplot(x = 'Pclass', y = 'Age', data=df, hue='Pclass', palette='tab10', estimator=np.std, errorbar=None)

# errorbar : 샘플로 추정한 모집단 추정치에 대한 오류 범위. CI 확신구간(Confidence Interval)
# estimator : 실제 value를 나타낼 방식
```

```python
# 등급별 생존률
sns.barplot(x = 'Pclass', y = 'Survived', data=df, hue='Pclass', palette='tab10', estimator=np.mean)#, errorbar=None)
```

```python
# 수평 바막대
sns.barplot(x= 'Survived', y = 'Pclass', data=df, hue='Pclass', estimator=np.sum, palette='tab10', orient='h')
```

```python
# hue 색상구분
# - x,y축 외의 데이터를 지정하면, 막대를 분리해(break down) 표현한다.

sns.barplot(x='Pclass', y='Survived', data=df, estimator=np.sum, hue= 'Sex', errorbar=None)
```

```python
# 성별에 따른 생존자 수를 구하고, hue를 Pclass로 하면?

sns.barplot(x='Sex', y = 'Survived', data=df, hue='Pclass', palette='tab10', estimator=np.mean, errorbar=None)
```

```python
# Age를 범주화 처리하기

# 성별/나이별 생존자 시각화
sns.barplot(x='Age', y='Survived', data=df, hue='Sex') # Age는 연속형 데이터라 굉장히 드러운 그래프가 나옴.
```

```python

def categroize_age(age) :
    match age // 10 :
        case 0: return '0대'
        case 1: return '10대'
        case 2: return '20대'
        case 3: return '30대'
        case 4: return '40대'
        case 5: return '50대'
        case 6: return '60대'
        case _: return '70대 이상'


df['AgeGroup'] = df['Age'].apply(categroize_age)

order_column = ['0대', '10대', '20대', '30대', '40대', '50대', '60대', '70대', '70대 이상']

df[['Age', 'AgeGroup']]

df_ = df[['PassengerId','Name','Age','AgeGroup','Sex','Survived']]
df_.dropna(how='any', inplace=True)
df_

sns.barplot(x='AgeGroup', y='Survived', data=df_, hue='AgeGroup', order=order_column) # 순서 지정 안하면 정규분포 방식나옴
```

```python
sns.barplot(x='Sex', y='Survived', data=df_, hue='AgeGroup', errorbar=None, hue_order=order_column)

plt.legend(loc='lower right')
```

## seaborn에서 subplot 작성하기
matplotlib의 subplot을 이용해서 axis별로 seaborn plot을 작성해야 한다.

```python
fig, ax = plt.subplots(1,3,figsize=(12,4))

#sns.countplot(x ='Survived', data=df,  palette='tab10', ax=ax[0])
#sns.countplot(x ='Pclass', data=df,  palette='tab10', ax=ax[1])
#sns.countplot(x ='Sex', data=df,  palette='tab10', ax=ax[2]) # 일일이 넣기

cate_column = ['Survived', 'Pclass', 'Sex']

for i, column in enumerate(cate_column) :
    print (i, column)
    sns.countplot(x = column, data = df, palette='tab10', ax = ax[i], hue = column)




```

```python
# 생사별, 등급별, 성별, 탑승항구별 승객 수
fig, ax = plt.subplots(2,2, figsize=(12,4))

catego_column = [['Survived','Pclass'],
                 ['Sex','Embarked']] # 2,2이기 때문에 그 틀에 맞춰서 해줘야 한다.

for i in range(len(catego_column)):
    for j in range(len(catego_column[i])):
        column = catego_column[i][j]
        sns.countplot(x= column, data=df, ax=ax[i][j], hue=column)
```

```python
fig, ax = plt.subplots(2,2, figsize=(8,8))

categories = np.array([
    ['Pclass', 'Sex'],
     ['Embarked', 'Parch']
])

# array에는 shape를 써야 한다.(원래는)

for i in range(categories.shape[0]):
    for j in range(categories.shape[1]):
        column = categories[i][j]
        sns.countplot(x= column, data=df, ax=ax[i][j], hue=column)

#for i in range(len(categories)):
 #   for j in range(len(categories[i])):
  #      column = categories[i][j]
   #     sns.(x= column, data=df, ax=ax[i][j], hue=column)
```

## heatmap
컬럼간의 상관도 correlation를 컬러 속성으로 시각화

컬럼데이터간의 상관계수를 계산하고, 1 ~ -1 사이로 표현한다. # 숫자들간의 관계
- 1 : 양의 상관관계(비례)
- 0 : 아무 관계 아님
- -1 : 음의 상관관계(반비례)

```python
# 상관계수 행렬 생성
#df.corr()
corr_matrix = df.corr(numeric_only=True) # 숫자 데이터를 가지고 있는   컬럼만 남겨둔다
corr_matrix
```

```python
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='YlGnBu_r')
```
